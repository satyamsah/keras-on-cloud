```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_x = np.linspace(-1,1,101)
```

```python
train_data_Y = 4 * train_x + np.random.rand(101)*.44
```

```python
train_x
```

```python
train_data_Y
```

```python
plt.scatter(train_x,train_data_Y,label='data',color=['red','green'] )
```

```python
model= Sequential()
```

```python
model.add(Dense(input_dim=1, activation="linear", units=1, kernel_initializer="uniform"))
```

```python
model.summary()
```

```python
len(model.layers)
```

```python
weights=model.layers[0].get_weights()
```

```python
len(weights)
```

```python
weights
```

```python
#intial weight
weights[0]
```

```python
#intial bias
weights[1]
```

```python
print("LR w and b",weights[0],weights[1])
```

```python
model.compile(optimizer='sgd',loss='mse')
```

```python
model.fit(train_x,train_data_Y,epochs=200,verbose=1)
```

```python
#final 
final_weights=model.layers[0].get_weights()
```

```python
final_weights[0]
```

```python
final_weights[1]
```

```python
result_train_x=model.predict(train_x)
```
